```python
nilai = [80, 75, 60, 90, 55, 70, 65, 88, 92, 50]
```

```python
maks = max(nilai)
mini = min(nilai)
```

```python
rata = sum(nilai) / len(nilai)

```

```python
lulus = sum(1 for n in nilai if n >= 60)
tidak_lulus = len(nilai) - lulus
print("Nilai:", nilai)
print("Tertinggi:", maks)
print("Terendah:", mini)
print("Rata-rata:", rata)
print("Jumlah lulus:", lulus)
print("Tidak lulus:", tidak_lulus)
```

```python
plt.figure()
plt.bar(["Tertinggi", "Terendah"], [maks, mini])
plt.title("Grafik Nilai Tertinggi dan Terendah")
plt.show()
```

```python
plt.figure()
plt.pie([lulus, tidak_lulus], labels=["Lulus", "Tidak Lulus"], autopct='%1.1f%%')
plt.title("Grafik Kelulusan")
plt.show()
```
